```python
#filling 5 to all the null values
df.fillna(5)
```

```python
#filling value to the partucular column's null values
df['name'].fillna('Sakthi')
```

```python
#replaces the NULL values with the values from the next row 
df.fillna(method='bfill')
```

```python
#replaces the NULL values with the value from the previous row 
df.fillna(method='ffill')
```

```python
# removing all the nul values
df.dropna()
```

```python
#will function like above if any value in a row is null then that row will be removed
df.dropna(how='any')
```

```python
#will remove if any value in a column is null then that column will be removed
df.dropna(how='any',axis='columns')
```

```python
# will remove columns if all the column values are null
df.dropna(how='all',axis='columns')
```

```python
#will remove the entire row if 'hsc' or 'SSLC' has any one value as null
df.dropna(how='any',subset=['hsc','SSLC'])
```

```python
df.fillna(method='bfill',inplace=True)
```

```python
# Finding duplicate values and by default the first occurance is Original and next occurance is duplicate
df.duplicated()
```

```python
#Make the first occurance as the true(original) and remaining occurances as duplicates
df.duplicated(keep='first')
```

```python
#Make the last occurance as the True and previous occurances as duplicates
df.duplicated(keep='last')
```

```python
#Make the all the duplicates as True
df.duplicated(keep=False)
```

```python
#dropping the duplicates occurances by keeping the first occurance 
df.drop_duplicates(keep='first')
```

```python
#dropping the duplicates occurances by keeping the last occurance 
df.drop_duplicates(keep='last')
```

```python
#dropping the duplicates occurances by keeping the first occurance of column names 'name' & 'college'. It will first remove the first column of df and next.
df.drop_duplicates(keep='first',subset=['name' or 'college'])
```
